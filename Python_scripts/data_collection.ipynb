```python
import serial

ser = serial.Serial('/dev/tty.usbmodem14401', 9600)
dataFile = open('FINAL_RIGHT_SLOUCH_WRONG_100.csv', 'w')
dataFile_ok = open('FINAL_RIGHT_SLOUCH_OK_100.csv', 'w')

dataFile.write('aX,aY,aZ,gX,gY,gZ\n')
dataFile_ok.write('aX,aY,aZ,gX,gY,gZ\n')

flag = 0
count = 0 

while True:
    if ser.in_waiting > 0:
        if count % 2 == 0:
            data = ser.readline().decode('utf-8').strip()
            dataFile.write(data + '\n')
            flag +=1
            if flag % 119 ==0:
                count = flag/119
                print("wrong: ", count)
                dataFile.write(' ' + '\n')

        if count % 2 != 0:
            data = ser.readline().decode('utf-8').strip()
            dataFile_ok.write(data + '\n')
            flag +=1
            if flag % 119 ==0:
                count = flag/119
                print("ok: ", count)
                dataFile_ok.write(' ' + '\n')
```
